# # Sari excel files ko ager ek sath kholna ho tou ye code use hoga

```python
df = pd.read_csv("./Sales_Data/Sales_April_2019.csv")

files = [file for file in os.listdir('./Sales_Data')]

all_months_data =pd.DataFrame()

for file in files:
    df = pd.read_csv("./Sales_Data/"+file)
    all_months_data = pd.concat([all_months_data, df])
    
all_months_data.to_csv("all_data.csv", index=False)
```

# find or and delete it 

```python
temp_df = all_data[all_data['Order Date'].str[0:2]!='Or']
temp_df.head()
```

```python
all_data = all_data[all_data['Order Date'].str[0:2]!= 'Or']
all_data.head()
```

# Convert data into correct type

```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])                # Make it 
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])                            #Make float
```

# Add Month Columb 

```python
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int32')
all_data.head()
```

# Task 3 = add a sales columb 

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()
```

# What is the best month of sales?How much was earned that month ?


```python
all_data.groupby('Month').sum()
```

```python
import matplotlib.pyplot as plt

months = range(1,13)
print(months)

plt.bar(months,all_data.groupby(['Month']).sum()['Sales'])
plt.xticks(months)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Month number')
plt.show()
```
